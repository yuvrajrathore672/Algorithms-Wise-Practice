```python
from sklearn.model_selection import train_test_split

from sklearn.datasets import make_classification
from sklearn.datasets import make_regression

from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import accuracy_score , classification_report
from sklearn.metrics import r2_score

from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import AdaBoostRegressor
```

```python
x,y = make_classification(
    n_features=20, 
    n_samples=1000, 
    n_informative=10,
    n_classes=2,
    random_state=42)

X_train , X_test ,y_train ,y_test = train_test_split(
    x,
    y,
    random_state=42, 
    test_size=0.2)
```

```python
# decision tree stump 
base_model = DecisionTreeClassifier(
    max_depth=1)                          # by defualt this already present in ada but for learning i am writing here

# b00sting 
abc = AdaBoostClassifier(
    base_model,
    n_estimators=200,
    learning_rate=0.1, 
    random_state=42
)

abc.fit(X_train,y_train)
```

```python
# prediction -- 

pred = abc.predict(X_test)

print(f"Accuracy score {accuracy_score(y_test,pred)}")
```

```python
# just for practice here using grid search tuning 

from sklearn.model_selection import GridSearchCV

para = {"n_estimators":[50,100,150,200,250,300]}

cv = GridSearchCV(
    estimator= AdaBoostClassifier(base_model,random_state=42),
    param_grid= para,
    cv=5, 
    scoring="accuracy"
)

cv.fit(X_train,y_train)
```

```python
print(f"Best parameters: {cv.best_params_}")
# print(cv.best_estimator_)

best_model = AdaBoostClassifier(n_estimators=200,random_state=42)
best_model.fit(X_train,y_train)
```

```python
p = best_model.predict(X_test)

print(accuracy_score(y_test,p))
```

#### Regressor ----------

```python
X , Y = make_regression(
    n_features=20,
    noise=0.1,
    n_samples=1000, 
    n_targets=1, 
    random_state=42
)

x_train , x_test , Y_train , Y_test = train_test_split(
    X,Y,random_state=42, test_size=0.2
)
```

```python
base_model_R  = DecisionTreeRegressor(random_state=42,max_depth=1)

abr = AdaBoostRegressor(random_state=42,estimator=base_model_R)

param_grid = {"n_estimators":[100,50,200,300],
             "learning_rate":[0.1,0.01,0.05,0.5]}

cvr = GridSearchCV(
    abr,
    param_grid= param_grid,cv=5, 
    scoring="r2"
)

cvr.fit(x_train,Y_train)
```

```python
cvr.best_params_
```

```python
best_model_R = AdaBoostRegressor(base_model_R,learning_rate=0.5,n_estimators=100)

best_model_R.fit(x_train,Y_train)
```

```python
pp = best_model_R.predict(x_test)

print(r2_score(Y_test,pp))
```
